```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rc
import numpy as np
import os
plt.style.use(['ggplot'])
rc('figure', dpi=256)
rc('font', family='serif', size=15)
rc('text',usetex=True)
```

```python
category_names = ['alg1-baseline', 'alg2-baseline']
funcs = ['sort', 'L2-norm', 'Heapify']
df_func_columns = []
for func in funcs:
    for category in category_names:
        df_func_columns.append(func)
```

```python
# with open(some file) as file:
# alg1 results sort= np.array(file.readline().split(" "))
# alg2 results sort= np.array(file.readline().split(" "))
```

```python
#random inputs for illustration purposes

l2_norm = np.array([
             np.random.randint(low=10, high=20, size=30), # alg1
             np.random.randint(low=10, high=20, size=30) # alg2
           ]).reshape((len(category_names), -1))

sort = np.array([
            np.random.randint(low=48, high=56, size=30), # alg1
            np.random.randint(low=20, high=30, size=30)  # alg2
           ]).reshape((len(category_names), -1))
heapify = np.array([
            [0] * 30,                                    # alg1
            np.random.randint(low=10, high=20, size=30)  # alg2
           ]).reshape((len(category_names), -1))
data = np.concatenate([l2_norm.T, sort.T, heapify.T], axis=1)
```

```python
df = pd.DataFrame(data, columns=df_func_columns)
tups = zip(*[category_names*len(data), df.keys()])
columns = pd.MultiIndex.from_tuples(tups, names=['algos', 'func'])
```

```python
df = pd.DataFrame(data, columns=columns)
mean = df.mean().sort_index(level=1)
columns = mean.index.get_level_values(1).unique().tolist() # order of column names changes when you sort index
mean = np.array(mean)
low_mean = np.array(mean - df.min()) # confidence intervals
high_mean = np.array(df.max() - mean)
```

```python
mean = mean.reshape((len(funcs), -1))
low_mean = low_mean.reshape((len(funcs), -1))
high_mean = high_mean.reshape((len(funcs), -1))
```

```python
results = {}
for items in list(zip(columns, mean)):
    results[items[0]] = items[1]

res_df = pd.DataFrame(results, index=category_names)
```

```python
ax = res_df.plot.barh(stacked=True, cmap='tab10', figsize=(16, 10), edgecolor='black', linewidth=1)
# 'Pastel1', 'Pastel2', 'Paired', 'Accent', 'Dark2', 'Set1', 'Set2', 'Set3', 'tab10', 'tab20', 'tab20b', 'tab20c'
for c in ax.containers:

    # format the number of decimal places and replace 0 with an empty string
    labels = [f'{w:.0f}' if (w := v.get_width()) > 0 else '' for v in c ]
    
    ax.bar_label(c, labels=labels, label_type='center')
    ax.set_title("Bottleneck analysis for $n = ...$")
    ax.set_xlabel("Runtime(ms)")

# curdir = os.getcwd() + "/"
# plt.tight_layout()
# plt.savefig(curdir + "bottlenecks")
```
